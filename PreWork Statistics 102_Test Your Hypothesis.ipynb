# <font color="red">I. Introduction to Hypothesis Testing </font>

**Hypothesis testing** in statistics is a way for you to test the results of an analysis or experiment to see if you have meaningful results. You’re basically testing whether your results are valid by figuring out the odds that your results have happened by chance. If your results may have happened by chance, the experiment won’t be repeatable and so has little use.<br>

Hypothesis tests use the following logic: “Given the human tendency to react to unusual but random behavior and interpret it as something meaningful and real, in our experiments we will require proof that the difference between groups is more extreme than what chance might reasonably produce.” This involves a baseline assumption that the treatments are equivalent, and any difference between the groups is due to chance. This baseline assumption is termed the null hypothesis. Our hope is then that we can, in fact, prove the null hypothesis wrong, and show that the outcomes for groups A and B are more different than what chance might produce.<br>

Statistical hypothesis testing plays an important role in the whole of statistics and in statistical inference. For example, Lehmann (1992) in a review of the fundamental paper by Neyman and Pearson (1933) says: "Nevertheless, despite their shortcomings, the new paradigm formulated in the 1933 paper, and the many developments carried out within its framework continue to play a central role in both the theory and practice of statistics and can be expected to do so in the foreseeable future".<br>

Significance testing has been the favored statistical tool in some experimental social sciences. Other fields have favored the estimation of parameters (e.g. effect size). Significance testing is used as a substitute for the traditional comparison of predicted value and experimental result at the core of the scientific method. When theory is only capable of predicting the sign of a relationship, a directional (one-sided) hypothesis test can be configured so that only a statistically significant result supports theory. This form of theory appraisal is the most heavily criticized application of hypothesis testing.

## Step 1: State Null and Alternative Hypotheses

A hypothesis is an approximate explanation that relates to the set of facts that can be tested by certain further investigations. There are basically two types, namely, null hypothesis and alternative hypothesis.<br>

The **null hypothesis** is generally denoted as H0. It states the exact opposite of what an investigator or an experimenter predicts or expects. It basically defines the statement which states that there is no exact or actual relationship between the variables.<br>

The **alternative hypothesis** is generally denoted as H1. It makes a statement that suggests or advises a potential result or an outcome that an investigator or the researcher may expect.

### One Sample Test:
Test this null hypothesis: the population mean for the treatment group is not significantly different from known or standard value c. This is stated as
 
H0: μ = c

The alternative hypothesis: the population mean is not equal to c or,
 
H1: μ ≠ c

### Two-sample Test:
Use the independent two-sample test when there is not a natural one-to-one pairing between the subjects in two treatment groups. The null hypothesis that the population means of the two groups are not significantly different: 
 
H0: μA = μB

The alternative hypothesis is 
 
H1: μA ≠ μB

### Example:
A gasoline additive is supposed to reduce the amount of carbon monoxide in automobile exhaust. A sample of automobiles are chosen. Half of them are given the additive; half of them are not. Test the null hypothesis that the amount of CO in the exhaust is not significantly different for the cars with the additive than it is for the cars without the additive.<br>

Please state the null hypothesis and alternative hypothesis:

H0: μA = μB not significantly different for the cars with the additive than it is for the cars without the additive.

The alternative hypothesis is

H1: μA ≠ μB significantly different for the cars with the additive than it is for the cars without the additive.

## Step 2: Verify Assumptions and Determine the Correct Test

Specific conditions or assumptions should be verified when we decide which test to conduct for our hypothesis testing, for example:

For Significance Test for a **Mean**:
* The sampling method is simple random sampling.
* The sample is drawn from a normal or near-normal population.<br>

For Significance Test for a **Proportion**:
* The sampling method is simple random sampling.
* Each sample point can result in just two possible outcomes. We call one of these outcomes a success and the other, a failure.
* The sample includes at least 10 successes and 10 failures.
* The population size is at least 10 times as big as the sample size.

**One Tailed vs Two Tailed**: 
Firstly one must decide whether the test has to be one tailed or two tailed. The statistical definitions of one tailed and two tailed tests are quite difficult. Hence, explanation by an example should serve the purpose.



Let us assume that we are comparing the differences between the Average Handling Times (AHT’s) of two different call centres. A two tailed test will check if there is any significant statistical difference in the samples being measured. This means that if one of the samples is significantly higher or significantly lower than the other, a difference will be shown. Hence in a two tailed test we are concerned about differences arising on both sides.

However, in the case of one tailed test we will first have to decide whether we want to compare for upper tail test or lower tailed test. The upper tailed test will check if one of the samples is significantly higher than the other. If the sample has a lower value, the null hypothesis will be selected and no difference will be shown. The exact opposite of this is the lower tailed test where null hypothesis will be rejected only if one sample is markedly lower than the other.

### Decision Criteria
Apart from the above there are three simple decision criteria upon which the selection of the correct hypothesis test is based. They are as follows:

1. **The Number of Groups Being Tested**: There are different hypothesis tests available if the statistical difference has to be checked for three samples and that of two samples. Clarifying the number of different samples will be the first step in selecting the correct test.

2. **Whether the Y’s are Discrete or Continuous**: There are different hypothesis tests available for discrete and continuous variables.

3. **Population Parameter Being Compared**: There are different hypothesis tests available for means, medians, standard deviations and even population parameters. Depending on what is being evaluated for difference, different tests may have to be used.

## Step 3: Determine the Confidence Interval and Level of Significance

In hypothesis testing, the **significance level** is the criterion used for rejecting the null hypothesis. In other words, it is the numerical measure that we use to determine the strength of the sample evidence
we are willing to consider strong enough to reject H0 is called the level of significance and it is denoted by the Greek letter alpha (α).

Traditionally, experimenters have used either the 0.05 level (sometimes called the 5% level) or the 0.01 level (1% level), although the choice of levels is largely subjective. The lower the significance level, the more the data must diverge from the null hypothesis to be significant. Therefore, the 0.01 level is more conservative than the 0.05 level.



Alpha levels can be controlled by you and are related to confidence levels. 

To get α, subtract your confidence level from 1:

* **For one-tail tests**, if you want to be 95 percent confident that your analysis is correct, the alpha level would be 1 – .95 = 5 percent.
* **For two-tailed tests**, divide the alpha level by 2. In this example, the two tailed alpha would be .05/2 = 2.5 percent. 

## Step 4: Calculate the Test Statistic and P-Value

### Test Statistic

A **test statistic** is a statistic (a quantity derived from the sample) used in statistical hypothesis testing. It is intuitively, how far away are your results, after a statistical formula, from the expected middle of statistically no change. Think of it as a form of standard deviation. The farther away a test statistic is away from the middle, the more likely it has a larger probability of being truely abnormal or enough evidence to say the null hypothesis is wrong.

### P-Value

The **P-value** is the probability of observing a sample statistic as extreme as the test statistic. 

P-values are from 0% to 100%. They are usually written as a decimal. For example, a p value of 5% is 0.05. Low p-values are good; They indicate your data did not occur by chance. For example, a p-value of .01 means there is only a 1% probability that the results from an experiment happened by chance. In most cases, a p-value of 0.05 (5%) is accepted to mean the data is valid.

#### Example 2

```python
## Two sample t test

import numpy as np

# Define 2 random distributions
N = 100
a = np.random.randn(N) + 2
b = np.random.randn(N)

# Calculate the Standard Deviation
var_a = a.var(ddof=1)
var_b = b.var(ddof=1)
s = np.sqrt((var_a + var_b)/2)

# Calculate the t-statistics
t = (a.mean() - b.mean())/(s*np.sqrt(2/N))

# Compare with the critical t-statistic
df = 2*N - 2  # Degree of Freedom

#p-value after comparison with the t 
p = 1 - stats.t.cdf(t,df=df)

print("t = " + str(t))
print("p = " + str(2*p))

# Cross Checking with the internal scipy function
t2, p2 = stats.ttest_ind(a,b)
print("t = " + str(t2))
print("p = " + str(p2))
```

#### Example 3

```python
def abtest(mu, lower, upper, lowervisit, uppervisit):
    x1 = np.random.randint(lower,upper,1)
    x2 = np.random.randint(lower,upper,1)
    n1 = np.random.randint(lowervisit,uppervisit,1)
    n2 = np.random.randint(lowervisit,uppervisit,1)
    Sp = upper - lower
    t = ((x1 - x2) - (mu)) / (Sp*((1/n1)+(1/n2))**.5)
    print(t, 'is your test statistic.')
    from scipy import stats
    p = 1 - stats.t.cdf(abs(t), df = n1 + n2 - 1)
    print(p, 'is your p-value.')

abtest(0, 20000, 30000, 100, 200)
```

## Step 5: Interpret Results and Make Conclusions

There are two ways you can evaluate your results:
1. Test Statistic
    1. If your test statistic is outside of the upper & lower limits of your confidence level, <br>
        then you state your alternative hypothesis (rejecting the null hypothesis).
    2. If your test statistic is inside the upper & lower limits, <br>
        then you state your null hypothesis (fail to reject the null hypothesis).
1. P-Value
    1. If your p-value is less than the limit of your confidence level, <br>
        then you state your alternative hypothesis (rejecting the null hypothesis).
    2. If your p-value is more than the limit, <br>
        then you state your null hypothesis (fail to reject the null hypothesis).   

For example, in the Example 2 above from Step 4, we can see that after comparing the t statistic with the critical t value, we get a good p value of 0.0005 and thus we reject the null hypothesis and thus it proves that the mean of the two distributions are different and statistically significant.<br>

Let's look at another example of results interpretation:

```python
def abtest(mu, lower, upper, lowervisit, uppervisit, confidencelevel):
    x1 = np.random.randint(lower,upper,1)
    x2 = np.random.randint(lower,upper,1)
    n1 = np.random.randint(lowervisit,uppervisit,1)
    n2 = np.random.randint(lowervisit,uppervisit,1)
    Sp = upper - lower
    t = ((x1 - x2) - (mu)) / (Sp*((1/n1)+(1/n2))**.5)
    print(t, 'is your test statistic.')
    from scipy import stats
    p = 1 - stats.t.cdf(abs(t), df = n1 + n2 - 1)
    print(p, 'is your p-value.')
    plimit = 1 - confidencelevel
    uppert = stats.t.ppf(confidencelevel, df = n1 + n2 - 1)
    lowert = stats.t.ppf(1 - confidencelevel, df = n1 + n2 - 1)
    if (t > uppert) or (t < lowert):
        print('Reject the null hypothesis based on the test statistic.')
    else:
        print('Failed to reject the null hypothesis based on test statistic.')
    if p < plimit:
        print('Reject the null hypothesis based on the p-value.')
    else:
        print('Failed to reject the null hypothesis based on p-value.')
    return 'Please state the proper hypothesis for conclusion.'
```

```python
abtest(200, 20000, 30000, 100, 200, .95)
```

She made a presentation to the Chief Operating Officer and the Board of Trustees and tried to convince them that CASH COW should invest in her 100 companies because they had made a significant more amount than their current 100 companies. Does the evidence prove her correct?

### Q.1 What is the null hypothesis ?

Null Hypothesis (H0)
H0: μ ≤ 6.62 Sasha's comp. mean is less than CASH COW's comp. mean

### Q.2 What is the alternative hypothesis ?

H1: μ > 6.62 Ssha's comp. mean is greater than CASH COW's comp. mean

### Q.3 Is this one-tail or two-tail ?

### Q.5 What is the p-value ?

```python
from scipy.stats import norm

x_bar = 7.08  # Sasha's comp.
mu = 6.62  # CASH COW's comp
sigma = 1.24  # Sd
n = 100  # sample size

# Test statistic
Z = (x_bar - mu) / (sigma / (n ** 0.5))

# P-value for one-tailed test
p_value = 1 - norm.cdf(abs(Z))
Z, p_value
```
